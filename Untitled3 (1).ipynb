# Mergesort

```python
def mergesort(arr,start,end) :
  kc=0
  if start==end:
    return kc
  mid=1+start+int((end-start)/2)
  kc+=mergesort(arr,start,mid-1)
  kc+=mergesort(arr,mid,end);
  kc+=merger(arr,start,mid,end)
  return kc

def merger(arr, start, mid, end) :
  temp = [0] * (end-start+1)
  x=start;
  y=mid;
  z=0;
  kc=0
  while(x<mid and y<end+1) :
    kc+=1
    if(arr[x]<arr[y]) :
      temp[z]=arr[x]
      x+=1
      z+=1
    else :
      temp[z]=arr[y]
      z+=1
      y+=1
  while(x<mid) :
    temp[z]=arr[x]
    x+=1
    z+=1
  while(y<end+1) :
    temp[z]=arr[y]
    y+=1
    z+=1
  for k in range (0,len(temp)):
    arr[start+k]=temp[k]
  return kc



```

```python
arr=[1,2,3,90,6,10,0]
kc=mergesort(arr,0,len(arr)-1);
print(arr)
print(kc)
```

# Insertion Sort

```python
def insertionsort(arr):
  n=len(arr);
  kc=0
  for h in range(1,n):
    for i in range(h,0,-1):
      kc+=1
      if(arr[i]<arr[i-1]):
        temp=arr[i-1]
        arr[i-1]=arr[i]
        arr[i]=temp
      else:
        break
  return kc




```

```python
arr2=[1,2,3,90,6,10,0]
kc=insertionsort(arr2)
print(arr2)
print(kc)
```

# Hybrid Merge-Insertion Sort

```python
def insertionsort2(arr,start,end):
  n=end-start+1;
  kc=0
  for h in range(1,n):
    for i in range(h,0,-1):
      kc+=1
      if(arr[start+i]<arr[start+i-1]):
        temp=arr[start+i-1]
        arr[start+i-1]=arr[start+i]
        arr[start+i]=temp
      else:
        break
  return kc
```

```python
def merger2(arr, start, mid, end) :
  temp = [0] * (end-start+1)
  x=start
  y=mid
  z=0
  kc=0
  while(x<mid and y<end+1) :
    kc+=1
    if(arr[x]<arr[y]) :
      temp[z]=arr[x]
      x+=1
      z+=1
    else :
      temp[z]=arr[y]
      z+=1
      y+=1
  while(x<mid) :
    temp[z]=arr[x]
    x+=1
    z+=1
  while(y<end+1) :
    temp[z]=arr[y]
    y+=1
    z+=1
  for k in range (0,len(temp)):
    arr[start+k]=temp[k]
  return kc
```

```python
def mergeinsertionsort(S,arr,start,end):
  kc=0
  if end-start+1<=S:
    kc=insertionsort2(arr,start,end)
    return kc
  mid=1+start+int((end-start)/2)
  kc+=mergeinsertionsort(S,arr,start,mid-1)
  kc+=mergeinsertionsort(S,arr,mid,end);
  kc+=merger2(arr,start,mid,end)
  return kc
```

```python
arr3=[1,2,3,90,6,10,0]
kc=mergeinsertionsort(6,arr3,0,len(arr3)-1)
print(arr3)
print(kc)
```

## Generating random input data

```python
from numpy.random import randint
from numpy.random import seed
inputdataset={}
seed(8)
r=1000
while(r<10000001):
  values = randint(0, 10001, r)
  inputdataset[r]=values
  r*=10
```

# Part c) i) Number of key comparisons vs Input array size for a fixed threshold S

```python
S=5
TC={}
for size,inputlist in inputdataset.items():
  TC[size]=mergeinsertionsort(S,inputlist,0,len(inputlist)-1)
print(TC)
```

# Part c) i) Average number of key comparisons (taken avg over datasets of different sizes) vs threshold value S
